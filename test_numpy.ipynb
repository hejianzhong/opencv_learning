```python
c1, c2, c3 = cv2.split(im)
plt.imshow(im)
```

```python
c1[10:200,] = 255
c2[:,10:200] = 255
img2 = cv2.merge((c1, c2, c3))
plt.imshow(img2)
```

```python
line_img = np.zeros([480, 640, 3], dtype=np.uint8)
cv2.line(line_img, (10,20), (300,400), (0,0,255), 5, -1)
plt.imshow(line_img)
```

```python
ellipse_img = np.zeros([480, 640, 3], dtype=np.uint8)
cv2.ellipse(ellipse_img, (200,200), (100, 100), 0, 0,360, (0,0,255))
cv2.ellipse(ellipse_img, (200,200), (100, 50), 0, 0,180, (0,0,255))
cv2.ellipse(ellipse_img, (200,200), (100, 50), 30, 180,360, (0,0,255))

plt.imshow(ellipse_img)
```

```python
poly_img = np.zeros([480, 640, 3], dtype=np.uint8)
point_set = [np.array([(20,20), (30,30), (40,40), (50,50), (20,50)], dtype=np.int32)]
cv2.polylines(poly_img,point_set ,True, (0,255,0))
plt.imshow(poly_img)
cv2.fillPoly(poly_img, point_set, (0,0,255))
plt.imshow(poly_img)
```
